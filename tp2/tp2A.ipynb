```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import get_cmap
```

```python
image = np.array([[0, 1],
[1, 0]])
```

```python
cmp = get_cmap('gray')
plt.imshow(image, cmap=cmp, vmin=0, vmax=1)
plt.show()
```

```python
image_flattened = image.flatten()
print(image_flattened) # R ́esultat : [0, 1, 1, 0]
```

```python
noise = np.random.normal(0, 0.1, image.shape)
image_noisy = np.clip(image + noise, 0, 1)
```

```python
plt.imshow(image_noisy, cmap=cmp, vmin=0, vmax=1)
plt.show()
```

```python
zA = [1.5,0.2]
zB = [0.3,1.1]

qA1 = np.exp(zA[0])/(np.exp(zA[0])+np.exp(zA[1]))
qA2 = np.exp(zA[1])/(np.exp(zA[0])+np.exp(zA[1]))

qB1 = np.exp(zB[0])/(np.exp(zB[0])+np.exp(zB[1]))
qB2 = np.exp(zB[1])/(np.exp(zB[0])+np.exp(zB[1]))

print(qA1, qA2)

pA = [1, 0]
pB = [0, 1]

eccA = -(pA[0]*np.log(qA1) + pA[1]*np.log(qA2))
eccB = -(pB[0]*np.log(qB1) + pB[1]*np.log(qB2))

print(eccA, eccB)
```
